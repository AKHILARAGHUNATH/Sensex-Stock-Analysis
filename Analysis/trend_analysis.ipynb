```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Load data
df = pd.read_excel(r"D:\Akhila\ADQ\SENSEX\VERSION_4\Datasets\Sensex_with weekends(1981 to 27-06-2025).xlsx")

# Clean column names and types
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')

# Extract components
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.strftime('%b')
df['Day'] = df['Date'].dt.day
df['Month_Num'] = df['Date'].dt.month

# Sort month names correctly
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Dropdowns
month_dropdown = widgets.Dropdown(
    options=month_order,
    value='Jan',
    description='Month:'
)

year_dropdown = widgets.Dropdown(
    options=sorted(df['Year'].unique().tolist()),
    value=1981,
    description='Year:'
)

# Callback
def show_month_year_data(month, year):
    filtered = df[(df['Month'] == month) & (df['Year'] == year)]
    daily_avg = filtered.groupby('Day')['Close'].mean().reset_index()
    daily_avg.columns = ['Day', 'Average Close']
    display(daily_avg)

# Interact
widgets.interact(show_month_year_data, month=month_dropdown, year=year_dropdown)
```
